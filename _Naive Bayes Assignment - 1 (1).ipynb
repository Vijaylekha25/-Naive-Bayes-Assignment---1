Q1. What is Bayes' theorem?

Answer- Bayes' theorem is a fundamental concept in probability theory, named after Thomas Bayes. It provides a way to update the probability of a hypothesis (or event) based on new evidence or information.

Q3. How is Bayes' theorem used in practice?

Answer- Bayes' theorem is used in various fields such as statistics, machine learning, and artificial intelligence. It is applied in spam filtering, medical diagnosis, document classification, and more. In practice, it helps in making predictions or decisions based on available data and prior knowledge.

Q4. What is the relationship between Bayes' theorem and conditional probability?

Answer- Bayes' theorem relates to conditional probability in that it provides a framework to update prior probabilities based on new evidence. Conditional probability calculates the probability of an event given that another event has occurred, while Bayes' theorem gives a way to revise the probability of an event based on additional information.

Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?

Answer-The choice of which type of Naive Bayes classifier to use depends on the nature of the problem and the assumptions about the data. The three common types of Naive Bayes classifiers are:

Gaussian Naive Bayes: Assumes that features follow a normal distribution.
Multinomial Naive Bayes: Suitable for discrete features (e.g., word counts).
Bernoulli Naive Bayes: Appropriate when features are binary (e.g., presence or absence of a feature).
The selection is based on the type of features and their distribution in the dataset.
